### Imports

```python
%matplotlib inline

import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt

from scripts.KDE_methods import method_silverman, method_abramson, method_RVKDE, method_ERVKDE, MSE

import time # for timing main()
from os import listdir # for checking if a file already exists
import warnings # used to suppress numba warnings
```

### Parameters

```python
# Set Scientific Notation for Pandas
pd.set_option('display.float_format', '{:.6E}'.format)

# Suppress Numba warnings
warnings.simplefilter('ignore')


RUNS=10
NUM_SAMPLES=12800

DF_OUTPATH = "kde_results" # folder to save results
IMG_OUTPATH = f"{DF_OUTPATH}/kde_plots" # folder to save images (plots)

IMG_FILENAME = f"kde_{NUM_SAMPLES}" # base name of image file for plots
DF_FILENAME = f"kde_MSE" # csv filename for final result dataframe

RESULT_CSV = f"{DF_OUTPATH}/{DF_FILENAME}.csv"
```

### Main()

```python
def main(num_samples=1000, condition=None, dimension=2, \
         plot=True, save=False, img_name='kde_fig', outpath="kde_plots"):
    """
    Returns MSEs of sample densities for 4 KDE methods (Silverman, RVKDE, ERVKDE, Abramson)
    
    num_samples: number of samples to be generated
    condition: list(zip(means, covariances)) of generated samples; (if None, uses single guassian distribution)
    dimension: dimension of samples; (currently may only work for dimension=2)
    
    plot: If True, plots KDE densities
    save: If True, will save plots to f"{outpath}/{img_name}.png"
    outpath: If None, uses current directory
    """
    if (outpath == None):
        img_pathname = f"{img_name}.png"
    else:
        img_pathname = f"{outpath}/{img_name}.png"
    
    if condition == None:
        condition = [
            [[0,0], [[1,0],[0,1]]]
                    ]
    
    xy_range=[-4,4] # range of X,Y axes
    spacing=100 # number of points on X,Y axes to test
    
    # Create X, Y axes
    xs = np.linspace(*xy_range,spacing)
    ys = np.linspace(*xy_range,spacing)
    X,Y = np.meshgrid(xs, ys)
    tofit = np.array([[x,y] for y in ys for x in xs])
    
    # Generate data
    samples = []
    densities = 0
    for cond in condition:
        samples += [np.random.multivariate_normal(*cond, num_samples)]
        densities += stats.multivariate_normal.pdf(tofit, *cond)
        
    data = np.vstack(np.array(samples))
    true_density = np.vstack(np.array(densities))
    
    X_data, Y_data = data[:,0], data[:,1] # for data scatter plot

    
    # Get KDE Densities
    silverman_results = method_silverman(data, tofit)
    RVKDE_results = method_RVKDE(data, xs, ys, dimension)
    ERVKDE_results = method_ERVKDE(data, xs, ys, dimension)
    abramson_results = method_abramson(data, xs, ys, dimension)
    
    
    # Plot Densities and Data
    if ((plot==True) or (save==True)):
        
        silverman_results = silverman_results.reshape(100,100)
        RVKDE_results = RVKDE_results.reshape(100,100)
        ERVKDE_results = ERVKDE_results.reshape(100,100)
        abramson_results = abramson_results.reshape(100,100)
        true_density = true_density.reshape(100,100)
        
        method_data = {'Silverman': [X, Y, silverman_results],
                       'RVKDE': [X, Y, RVKDE_results],
                       'ERVKDE': [X, Y, ERVKDE_results],
                       'Abramson': [X, Y, abramson_results],
                       'Distribution': [X,Y,true_density],
                       'Data': [X_data,Y_data],
                      }
        
        method_names = list(method_data.keys())

        plt.figure(figsize=(20,10))

        jump = 0 # used to move plots to next line
        for i in range(len(method_names)):
            method_name = method_names[i]
            ax = plt.subplot(f'24{i+1+jump}')

            if method_name == 'Data':
                plt.scatter(*method_data[method_name],alpha=0.5) # alpha=0.1
#                 jump += 2
            else:
                ax.contourf(*method_data[method_name])

            ax.set_title(f"{method_name}")
            ax.set(xlabel="X", ylabel="Y")
            ax.set_facecolor('#44146f')

        plt.tight_layout(pad=1.0)
        if save == True:
            plt.savefig(img_pathname)
            
        if (plot == True):
            plt.show()
        else:
            plt.close()
            
        
    silverman_results = silverman_results.flatten()
    RVKDE_results = RVKDE_results.flatten()
    ERVKDE_results = ERVKDE_results.flatten()
    abramson_results = abramson_results.flatten()
    true_density = true_density.flatten()
    

    # Get Mean Squared Error
    MSE_silverman = MSE(true_density, silverman_results)
    MSE_RVKDE = MSE(true_density, RVKDE_results)
    MSE_ERVKDE = MSE(true_density, ERVKDE_results)
    MSE_abramson = MSE(true_density, abramson_results)
    
    MSEs = [MSE_silverman, MSE_RVKDE, MSE_ERVKDE, MSE_abramson]
    
    return MSEs
    
```

```python
# RUNS=10 # <-- Set in parameters section
# NUM_SAMPLES=1000 # <-- Set in parameters section

# Note: The first run is slower due to how Numba JIT compilation works


results_array = []

total_time = time.time()

# Get average results over runs
for run in range(RUNS):
    t = time.time()
    
    if (run == 0):
        result = main(num_samples=NUM_SAMPLES, plot=True, save=True, \
                      img_name=f'{IMG_FILENAME}_{run}', outpath=IMG_OUTPATH)
    
    else:
        result = main(num_samples=NUM_SAMPLES, plot=False, save=False)
    
    t = time.time() - t
    print(f"{run}: main complete -  {t:.3f} seconds")
    
    results_array += [result]
    
total_time = time.time() - total_time
total_time = round(total_time, 3)

result_df = pd.DataFrame(results_array, columns=["Silverman", "RVKDE", "ERVKDE", "Abramson"])

print(f"\n Time elapsed: {total_time} seconds")
display(result_df)

result_df.mean()
```

### Save MSE Results

```python
# Save average MSE results to dataframe
final_df = pd.DataFrame([result_df.mean()])
final_df["Runs"] = RUNS
final_df["Samples"] = NUM_SAMPLES
final_df["Time"] = str(total_time)
final_df = final_df.reindex(columns=["Samples", "Runs", "Time", "Silverman", "RVKDE", "ERVKDE", "Abramson"])

# Update RESULT_CSV with average MSE results
if RESULT_CSV.split("/")[-1] in listdir(DF_OUTPATH):
    # RESULT_CSV already exists
    stored_results = pd.read_csv(RESULT_CSV)
    stored_results = stored_results.append(final_df)
    stored_results.to_csv(RESULT_CSV, index=False)
else:
    # RESULT_CSV does not exist yet
    final_df.to_csv(RESULT_CSV, index=False)
    stored_results = final_df

stored_results
```
